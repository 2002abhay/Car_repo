```python
def convert_to_int(string):
    # Remove the '₹' sign, commas, and any non-numeric characters
    cleaned_string = re.sub(r'[₹,]', '', string)
    
    # Initialize the multiplier
    multiplier = 1
    
    # Check if the string contains 'Lakh' or 'Crore' and convert to the appropriate multiplier
    if 'Lakh' in cleaned_string:
        cleaned_string = cleaned_string.replace(' Lakh', '')  # Update the string
        multiplier = 100000
    elif 'Crore' in cleaned_string:
        cleaned_string = cleaned_string.replace(' Crore', '')  # Update the string
        multiplier = 10000000

    # Convert the cleaned string to a float and then to an integer by multiplying with the appropriate multiplier
    numeric_value = float(cleaned_string)
    result = int(numeric_value * multiplier)

    return result
```

```python
def extract_mileage(mileage_array):
    mileage_values = []
    for mileage_str in mileage_array:
        # Use regular expression to find and extract the numeric part
        match = re.search(r'[\d.]+', mileage_str)
        if match:
            mileage_values.append(float(match.group()))
        else:
            mileage_values.append(None)  # You can choose to handle invalid formats differently

    return mileage_values
```
